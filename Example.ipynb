```python
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
import tensorflow as tf

from matplotlib import rcParams
import matplotlib.pyplot as plt
import pandas as pd
import nilmtk
from nilmtk import DataSet, MeterGroup

import os
from nilmtk.datastore import HDFDataStore

import metrics

import csv

print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
print(tf.test.is_gpu_available())
print(tf.config.experimental.list_physical_devices('GPU'))

plt.style.use('ggplot')
rcParams['figure.figsize'] = (18, 18)

path = '/home/nick/Documents/ukdale.h5'
uk =  DataSet(path)
elec1 = uk.buildings[1].elec
```

## Train on 1

```python
train = uk
train.set_window(start="2013-01-01", end="2013-10-01")
train_elec = elec1
```

## Train Model

```python
epochs = 5
iterations = 3
batch_size = 1024
r1 = []
r2 = []
r3 = []
r4 = []
r5 = []
for i in range(iterations):
    print('##########################################################')
    print('##### {0} #####'.format(i))
    print('##########################################################')
    from windowgrudisaggregatorAdd import WindowGRUDisaggregator
    dev = 'microwave'
    meter_key = 'microwave'
    model_name = dev+'iter'+str(i)+'_self_uk.h5'
    fn = dev+'_model_self.csv'
    gru = WindowGRUDisaggregator(window_size=50)
    train_mains = elec1.mains()
    train_meter = train_elec.submeters()[meter_key]
    gru.train(train_mains, train_meter,batch_size=batch_size, epochs=epochs, sample_period=6)
    
    ## Test on HOUSE 1
    print('##########################################################')
    print('TEST HOUSE 1  start=2013-10-01, end=2014-01-01')
    print('##########################################################')
    b = 1
    test = DataSet(path)
    test.set_window(start="2013-10-01", end="2014-01-01")
    test_elec = test.buildings[b].elec
    test_mains = test_elec.mains()

    disag_filename = dev+str(b)+'iter'+str(i)+'_self_uk.h5' # The filename of the resulting datastore
    output = HDFDataStore(disag_filename, 'w')

    gru.disaggregate(test_mains, output, train_meter, sample_period=6)
    result = DataSet(disag_filename)
    res_elec = result.buildings[b].elec
    predicted = res_elec[meter_key]
    ground_truth = test_elec[meter_key]

    import matplotlib.pyplot as plt
    predicted.plot()
    ground_truth.plot()
    plt.show()

    rpaf = metrics.recall_precision_accuracy_f1(predicted, ground_truth)

    print("Recall: {}".format(rpaf[0]))
    print("Precision: {}".format(rpaf[1]))
    print("Accuracy: {}".format(rpaf[2]))
    print("F1 Score: {}".format(rpaf[3]))

    print("Relative error in total energy: {}".format(metrics.relative_error_total_energy(predicted, ground_truth)))

    print("Mean absolute error(in Watts): {}".format(metrics.mean_absolute_error(predicted, ground_truth)))
    
    rpaf2 = rpaf
    rpaf2 = list(rpaf2)
    rpaf2.append(metrics.relative_error_total_energy(predicted, ground_truth))
    rpaf2.append(metrics.mean_absolute_error(predicted, ground_truth))
    rpfa2 = tuple(rpaf2)
    r1.append(rpaf2)
    
    with open(fn, mode='w') as csv_file:
        fieldnames = ['train_House','test_House', 'Recall', 'Precision','Accuracy','F1', 'RETE', 'METE']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerow({'train_House': '1', 'test_House': b, 'Recall': rpaf[0], 
                         'Precision': rpaf[1],'Accuracy': rpaf[1],'F1': rpaf[3],
                         'RETE': metrics.relative_error_total_energy(predicted, ground_truth), 
                         'METE': metrics.mean_absolute_error(predicted, ground_truth)})

    output.close()
    
    print('##########################################################')
    print('TEST HOUSE 2  start=2013-06-22, end=2013-09-22')
    print('##########################################################')
    ## Test on HOUSE 2
    b = 2
    test = DataSet(path)
    test.set_window(start="2013-06-22", end="2013-09-22")
    test_elec = test.buildings[b].elec
    test_mains = test_elec.mains()

    disag_filename = dev+str(b)+'iter'+str(i)+'_self_uk.h5' # The filename of the resulting datastore
    output = HDFDataStore(disag_filename, 'w')

    gru.disaggregate(test_mains, output, train_meter, sample_period=6)
    result = DataSet(disag_filename)
    res_elec = result.buildings[b].elec
    predicted = res_elec[meter_key]
    ground_truth = test_elec[meter_key]

    import matplotlib.pyplot as plt
    predicted.plot()
    ground_truth.plot()
    plt.show()

    rpaf = metrics.recall_precision_accuracy_f1(predicted, ground_truth)
    
    print("Recall: {}".format(rpaf[0]))
    print("Precision: {}".format(rpaf[1]))
    print("Accuracy: {}".format(rpaf[2]))
    print("F1 Score: {}".format(rpaf[3]))

    print("Relative error in total energy: {}".format(metrics.relative_error_total_energy(predicted, ground_truth)))

    print("Mean absolute error(in Watts): {}".format(metrics.mean_absolute_error(predicted, ground_truth)))
    
    rpaf2 = rpaf
    rpaf2 = list(rpaf2)
    rpaf2.append(metrics.relative_error_total_energy(predicted, ground_truth))
    rpaf2.append(metrics.mean_absolute_error(predicted, ground_truth))
    rpfa2 = tuple(rpaf2)
    r2.append(rpaf2)
    
    with open(fn, mode='w') as csv_file:
        fieldnames = ['train_House','test_House', 'Recall', 'Precision','Accuracy','F1', 'RETE', 'METE']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerow({'train_House': '1', 'test_House': b, 'Recall': rpaf[0], 
                         'Precision': rpaf[1],'Accuracy': rpaf[1],'F1': rpaf[3],
                         'RETE': metrics.relative_error_total_energy(predicted, ground_truth), 
                         'METE': metrics.mean_absolute_error(predicted, ground_truth)})

    output.close()
    
    # Test on house 4
    print('##########################################################')
    print('TEST HOUSE 4 start="2013-03-09", end="2013-06-09"')
    print('##########################################################')
    b = 4
    test = DataSet(path)
    test.set_window(start="2013-03-09", end="2013-06-09")
    test_elec = test.buildings[b].elec
    test_mains = test_elec.mains()

    disag_filename = dev+str(b)+'iter'+str(i)+'_self_uk.h5' # The filename of the resulting datastore
    output = HDFDataStore(disag_filename, 'w')

    gru.disaggregate(test_mains, output, train_meter, sample_period=6)
    result = DataSet(disag_filename)
    res_elec = result.buildings[b].elec
    predicted = res_elec[meter_key]
    ground_truth = test_elec[meter_key]

    predicted.plot()
    ground_truth.plot()
    plt.show()

    rpaf = metrics.recall_precision_accuracy_f1(predicted, ground_truth)
    
    print("Recall: {}".format(rpaf[0]))
    print("Precision: {}".format(rpaf[1]))
    print("Accuracy: {}".format(rpaf[2]))
    print("F1 Score: {}".format(rpaf[3]))

    print("Relative error in total energy: {}".format(metrics.relative_error_total_energy(predicted, ground_truth)))

    print("Mean absolute error(in Watts): {}".format(metrics.mean_absolute_error(predicted, ground_truth)))
    
    rpaf2 = rpaf
    rpaf2 = list(rpaf2)
    rpaf2.append(metrics.relative_error_total_energy(predicted, ground_truth))
    rpaf2.append(metrics.mean_absolute_error(predicted, ground_truth))
    rpfa2 = tuple(rpaf2)
    r3.append(rpaf2)
    
    with open(fn, mode='a') as csv_file:
        fieldnames = ['train_House','test_House', 'Recall', 'Precision','Accuracy','F1', 'RETE', 'METE']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writerow({'train_House': '1', 'test_House': b, 'Recall': rpaf[0], 
                         'Precision': rpaf[1],'Accuracy': rpaf[1],'F1': rpaf[3],
                         'RETE': metrics.relative_error_total_energy(predicted, ground_truth), 
                         'METE': metrics.mean_absolute_error(predicted, ground_truth)})

    output.close()
    
    
    print('##########################################################')
    print('TEST HOUSE 5 start=2014-07-15, end=2014-10-15')
    print('##########################################################')
    # Test on House 5
    b = 5
    test = DataSet(path)
    test.set_window(start='2014-07-15', end='2014-10-15')
    test_elec = test.buildings[b].elec
    test_mains = test_elec.mains()

    disag_filename = dev+str(b)+'iter'+str(i)+'_self_uk.h5' # The filename of the resulting datastore
    output = HDFDataStore(disag_filename, 'w')

    gru.disaggregate(test_mains, output, train_meter, sample_period=6)
    result = DataSet(disag_filename)
    res_elec = result.buildings[b].elec
    predicted = res_elec[meter_key]
    ground_truth = test_elec[meter_key]

    predicted.plot()
    ground_truth.plot()
    plt.show()

    rpaf = metrics.recall_precision_accuracy_f1(predicted, ground_truth)
    
    print("Recall: {}".format(rpaf[0]))
    print("Precision: {}".format(rpaf[1]))
    print("Accuracy: {}".format(rpaf[2]))
    print("F1 Score: {}".format(rpaf[3]))

    print("Relative error in total energy: {}".format(metrics.relative_error_total_energy(predicted, ground_truth)))

    print("Mean absolute error(in Watts): {}".format(metrics.mean_absolute_error(predicted, ground_truth)))
    rpaf2 = rpaf
    rpaf2 = list(rpaf2)
    rpaf2.append(metrics.relative_error_total_energy(predicted, ground_truth))
    rpaf2.append(metrics.mean_absolute_error(predicted, ground_truth))
    rpfa2 = tuple(rpaf2)
    r5.append(rpaf2)
    
    with open(fn, mode='a') as csv_file:
        fieldnames = ['train_House','test_House', 'Recall', 'Precision','Accuracy','F1', 'RETE', 'METE']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writerow({'train_House': '1', 'test_House': b, 'Recall': rpaf[0], 
                         'Precision': rpaf[1],'Accuracy': rpaf[1],'F1': rpaf[3],
                         'RETE': metrics.relative_error_total_energy(predicted, ground_truth), 
                         'METE': metrics.mean_absolute_error(predicted, ground_truth)})

    output.close()
    del gru

```
